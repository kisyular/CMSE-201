```python
fireReport = pd.read_csv('https://raw.githubusercontent.com/kisyular/CMSE-201/master/NFPA-23688.csv')
fireReport
fireCase = pd.read_csv("https://github.com/kisyular/CMSE-201/blob/master/fireReports.xlsx?raw=true")
```

   **1. Examine the trend across 1980 to 2011 and show the variation of the number of people who performed  firefighting as Volunteer and as a Career. Explain the relationship between the two**

```python
y = fireReport["Year"] # Slice the dataframe to get the years, Volunteer Fire Fighters and Career Fire Fighters
volunteer = fireReport["Volunteer firefighters"]
career = fireReport["Career firefighters"]
x =[]
year = []
for element in y:
    x =element.split("-") # Split the year string into three indices
    year.append(int(x[0])) # Append the firts index as an int to the Year List

plt.plot(year, volunteer, linewidth = 3, linestyle="solid", marker="*", color="green", label = "Volunteer firefighters")
plt.plot(year, career, linewidth = 3, linestyle="solid", marker="*", color="red", label = "Career firefighters")

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.xlabel("Time in Years")
plt.ylabel("Fire Fighters")
plt.title("Fire Fighter as Volunteer and as Career  vs Years of Service")
```

<font color='green'>As seen from the graph, the number of people who participated or performed fire fighting are increasing if they perform it as a Career but decreasing when they perform it as a Volunteer service. </font>

```python
total = fireReport["Total"]
plt.plot(year, volunteer, linewidth = 3, linestyle="solid", marker="*", color="green", label = "Volunteer firefighters")
plt.plot(year, total, linewidth = 3, linestyle="solid", marker="*", color="red", label = "Total Participation")

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.xlabel("Time in Years")
plt.ylabel("Fire Fighters")
plt.title("Fire Fighter as Volunteer and Total Participation  vs Years of Service")
```
